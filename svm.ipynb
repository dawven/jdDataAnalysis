```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model,cross_validation,svm
import pandas as pd
from pandas import DataFrame
from sklearn.preprocessing import MinMaxScaler

def load_data_classfication():
    df = pd.read_csv('c:\\Users\\Administrator\\Desktop\\data.csv',encoding = 'GBK')
    X = df.iloc[:500,:5]
    y = df.iloc[:500,8:]
    #X_test = df.iloc[500:1000,:5]
    #Y_test = df.iloc[500:1000,8:]
    X_train=X  #本句与下一句是告诉机器谁是X，谁是Y.如果不告诉模型谁是x 谁是y  模型很困扰，分类结果不准确。   
               #如果你的业务数据  已经分类x y  就不需要红框部分了; 否则需要指定。
    Y_train=y
    return cross_validation.train_test_split(X_train,Y_train,test_size=0.25,random_state=0,stratify=Y_train)

def test_SVC_linear(*data):
    X_train,X_test,Y_train,Y_test=data
    #对数据进行归一化
    min_max_scaler = MinMaxScaler()
    X_train = min_max_scaler.fit_transform(X_train)
    X_test = min_max_scaler.fit_transform(X_test)
    #训练分类器
    clf=svm.SVC(kernel='linear')
    clf.fit(X_train,Y_train)
    print('Coefficients:%s, intercept %s'%(clf.coef_,clf.intercept_))
    print('Score:%.2f' % clf.score(X_test,Y_test))
    #预测
    test_predict=clf.predict(X_test)
    #取分类结果
    label_list = list(test_predict)
    print (label_list)
    #预测准确率
    clf.score(X_test,Y_test)
    print(np.mean(test_predict==Y_test))
    print(Y_test)
    print(test_predict)
    #测试新数据集
    iris_test=pd.read_csv('c:\\Users\\Administrator\\Desktop\\svm.csv')
    X=iris_test[['cup','days','hour','mobile','productcolor']]
    Y=DataFrame(clf.predict(X),columns=['class'])
    print (X)
    print (Y)
    #把X_predict写入到iris_test.csv文件中。
    XY=pd.concat([X,Y],axis=1)
    print (XY)
    #Y=[X,X_predict]
    #print Y
    XY.to_csv('E:/aa.csv',index=[]) #index=[]是为了在写入文件时，不把索引写入

if __name__=="__main__":
    X_train,X_test,Y_train,Y_test=load_data_classfication()
    test_SVC_linear(X_train,X_test,Y_train,Y_test)
```
